```python
import numpy as np
import pandas as pd
import math
import random
import numbers
import functions
import functions.labeled_set as ls
import functions.decision_tree as dtree
import functions.perceptron as pt
from copy import deepcopy

from datetime import datetime as dt
import matplotlib.pyplot as plt
```

```python
match = pd.read_csv("soccer/match3.csv")
df_match = match
df_match[0:5]
```

```python
team_attributes = pd.read_csv("soccer/teamattributes.csv")
team_attributes[0:4]
```

```python
df_match1 = df_match[["home_team_api_id","away_team_api_id","home_team_goal","away_team_goal"]]
```

```python
def difference(df_match):
    "Calcul pour chaque équipe la différence entre le nombre de matchs gangnés et le nombre de match perdus"
    diff = dict()
    
    for i in range(len(df_match)):
        id_home = df_match.home_team_api_id.iloc[i]
        id_away = df_match.away_team_api_id.iloc[i]
        
        if df_match.home_team_goal.iloc[i] > df_match.away_team_goal.iloc[i]:
            
            if id_home in diff.keys(): 
                diff[id_home][0] += 1
                diff[id_home][1] += 1
            else: 
                diff[id_home] = [1,1]
                
            if id_away in diff.keys(): 
                diff[id_home][1] += 1
            else: 
                diff[id_away] = [0,1]
            
        elif df_match.home_team_goal.iloc[i] < df_match.away_team_goal.iloc[i]:
            
            if id_away in diff.keys():
                diff[id_away][0] += 1
                diff[id_away][1] += 1
            else:
                diff[id_away] = [1,1]
                
            if id_home in diff.keys():
                diff[id_home][1] += 1
            else:
                diff[id_home] = [0,1]
        
    return diff
```

```python
diff = difference(df_match1)
```

```python
def from_date_to_season(date):
    year = int(date[0:4])
    month = int(date[5:7])
    if month < 8 :
        return str(year-1) + '/' + str(year)
    else :
        return str(year) + '/' + str(year+1)
```

```python
def transform_team_attributes_df(df):
    columns = ['team_api_id', 'season', 'buildUpPlaySpeed',
               'buildUpPlayDribbling', 'buildUpPlayPassing',
               'buildUpPlayPositioningClass', 'chanceCreationPassing',
               'chanceCreationCrossing', 'chanceCreationShooting',
               'chanceCreationPositioningClass', 'defencePressure',
               'defenceAggression', 'defenceTeamWidth',
               'defenceDefenderLineClass']
    
    numeric_columns = ['buildUpPlaySpeed', 'buildUpPlayDribbling',
                      'buildUpPlayPassing', 'chanceCreationPassing',
                      'chanceCreationCrossing', 'chanceCreationShooting',
                      'defencePressure', 'defenceAggression',
                      'defenceTeamWidth']
    
    new_df = dict()
    for i in reversed(columns):
        new_df[i] = []
    
    for i in range(len(df)):
        new_df['team_api_id'].append( df['team_api_id'].iloc[i] )
        new_df['season'].append( from_date_to_season(df['date'].iloc[i]) )
        for j in numeric_columns:
            if math.isnan(df[j].iloc[i]):
                new_df[j].append(17)
                #la seule colonne contenant des NaN est 'buildUpPlayDribbling'
                #et les NaN n'apparaissent que lorsque 'buildUpPlayDribblingClass'
                # est a 'Little'
            else:
                new_df[j].append(df[j].iloc[i])
        if df["buildUpPlayPositioningClass"].iloc[i] == "Organised":
            new_df["buildUpPlayPositioningClass"].append(1)
        else :
            new_df["buildUpPlayPositioningClass"].append(0)
       
        if df["chanceCreationPositioningClass"].iloc[i] == "Organised":
            new_df["chanceCreationPositioningClass"].append(1)
        else :
            new_df["chanceCreationPositioningClass"].append(0)
        
        if df["defenceDefenderLineClass"].iloc[i] == "Cover":
            new_df["defenceDefenderLineClass"].append(1)
        else :
            new_df["defenceDefenderLineClass"].append(0)
    
    return pd.DataFrame(new_df)
            
        
```

```python
df_team = transform_team_attributes_df(team_attributes)
df_team[0:5]
```

```python
def create_labeled_set(df_match, df_team,difference):
    """creer un labeled set a partir du datafram des matchs
        et du dataframe des caracteristiques des equipes
    """
    columns = ['buildUpPlaySpeed',
           'buildUpPlayDribbling', 'buildUpPlayPassing',
           'buildUpPlayPositioningClass', 'chanceCreationPassing',
           'chanceCreationCrossing', 'chanceCreationShooting',
           'chanceCreationPositioningClass', 'defencePressure',
           'defenceAggression', 'defenceTeamWidth',
           'defenceDefenderLineClass']

    
    dataset = ls.LabeledSet( 2 * len(columns) +2)

    for i in range(len(df_match)):
        season = df_match["season"].iloc[i]
        id1 = df_match["home_team_api_id"].iloc[i]
        id2 = df_match["away_team_api_id"].iloc[i]
        score1 = df_match["home_team_goal"].iloc[i]
        score2 = df_match["away_team_goal"].iloc[i]
        
        team1 = df_team.loc[(df_team['season'] == season) & (df_team['team_api_id'] == id1) ]
        team2 = df_team.loc[(df_team['season'] == season) & (df_team['team_api_id'] == id2) ]
        if(len(team1) > 0 and len(team2)>0):
            vector = []
            label = -1
            for j in columns:
                vector.append(team1[j].iloc[0])
            a = difference[id1][0]/difference[id1][1]
            vector.append(a)
            
            for j in columns:
                vector.append(team2[j].iloc[0])
            b = difference[id2][0]/difference[id2][1]
            vector.append(b)
            
            if score1 >= score2:
                label = 1
            
            dataset.addExample(vector,label)
    return dataset        
```

```python
labeledset = create_labeled_set(df_match[0:1000], df_team, diff)
train_set, test_set = ls.split(labeledset)
```

```python
nb_it = 200
kernel = pt.KernelPoly( labeledset.getInputDimension())
learning_rate = 0.0000001
perceptron = pt.PerceptronKernel(kernel.output_dimension(),learning_rate, kernel)

accuracy_test = []
accuracy_train = []
accuracy_all = []


for i in range(nb_it):
    accuracy_test.append(perceptron.accuracy(test_set)*100)
    accuracy_train.append(perceptron.accuracy(train_set)*100)
    accuracy_all.append(perceptron.accuracy(labeledset)*100)
    perceptron.train(train_set)

plt.figure()

plt.plot(range(nb_it),accuracy_test,label="test_set")
plt.plot(range(nb_it),accuracy_train,label="train_set")
plt.plot(range(nb_it),accuracy_all,label="all")
plt.legend(loc="best")
plt.title("Accuracy en fonction du nombre d'itérations")
plt.ylabel("Accuracy en pourcentage")
plt.xlabel("Nombre d'itération")
```
