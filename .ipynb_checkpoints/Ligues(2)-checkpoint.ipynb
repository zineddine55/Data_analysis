```python
team_attributes = pd.read_csv("soccer/teamattributes.csv")
team_attributes[0:4]
```

```python
def from_date_to_season(date):
    year = int(date[0:4])
    month = int(date[5:7])
    if month < 8 :
        return str(year-1) + '/' + str(year)
    else :
        return str(year) + '/' + str(year+1)
```

```python
def transform_team_attributes_df(df):
    columns = ['team_api_id', 'season', 'buildUpPlaySpeed',
               'buildUpPlayDribbling', 'buildUpPlayPassing',
               'buildUpPlayPositioningClass', 'chanceCreationPassing',
               'chanceCreationCrossing', 'chanceCreationShooting',
               'chanceCreationPositioningClass', 'defencePressure',
               'defenceAggression', 'defenceTeamWidth',
               'defenceDefenderLineClass']
    
    numeric_columns = ['buildUpPlaySpeed', 'buildUpPlayDribbling',
                      'buildUpPlayPassing', 'chanceCreationPassing',
                      'chanceCreationCrossing', 'chanceCreationShooting',
                      'defencePressure', 'defenceAggression',
                      'defenceTeamWidth']
    
    new_df = dict()
    for i in reversed(columns):
        new_df[i] = []
    
    for i in range(len(df)):
        new_df['team_api_id'].append( df['team_api_id'].iloc[i] )
        new_df['season'].append( from_date_to_season(df['date'].iloc[i]) )
        for j in numeric_columns:
            if math.isnan(df[j].iloc[i]):
                new_df[j].append(17)
                #la seule colonne contenant des NaN est 'buildUpPlayDribbling'
                #et les NaN n'apparaissent que lorsque 'buildUpPlayDribblingClass'
                # est a 'Little'
            else:
                new_df[j].append(df[j].iloc[i])
        if df["buildUpPlayPositioningClass"].iloc[i] == "Organised":
            new_df["buildUpPlayPositioningClass"].append(1)
        else :
            new_df["buildUpPlayPositioningClass"].append(0)
       
        if df["chanceCreationPositioningClass"].iloc[i] == "Organised":
            new_df["chanceCreationPositioningClass"].append(1)
        else :
            new_df["chanceCreationPositioningClass"].append(0)
        
        if df["defenceDefenderLineClass"].iloc[i] == "Cover":
            new_df["defenceDefenderLineClass"].append(1)
        else :
            new_df["defenceDefenderLineClass"].append(0)
    
    return pd.DataFrame(new_df)
```

```python
df_team = transform_team_attributes_df(team_attributes)
df_team[0:5]
```

```python
match = pd.read_csv("soccer/match.csv")
df_match = match[["league_id","home_team_api_id","away_team_api_id"]]
df_match[0:5]
```

```python
def teams_league(df_match):
    " determine les equipes appartenant a cahque ligue"
    leagues = dict()
    
    for i in df_match.league_id.unique():
        leagues[i] = set()
    
    for i in range(len(df_match)):
        leagues[df_match.league_id.iloc[i]].update([df_match.home_team_api_id.iloc[i],df_match.away_team_api_id.iloc[i]])
        
    return leagues
```

```python
leagues = teams_league(df_match)
for l in leagues.keys():
    print("league : ",l," equipes : ",len(leagues[l]))
```

```python
def league_dataframe(leagues, df_team):
    list_df=[]
    
    for i in leagues.keys():
        list_df.append(df_team.loc[df_team["team_api_id"].isin(leagues[i]) | df_team["season"].isin(['2014/2015'])])
        
    return list_df
```

```python
list_dfleagues = league_dataframe(leagues, df_team)
```

```python
list_dfleagues[1].drop(["season","team_api_id"],1)[0:5]
```

```python
def leagues_to_df(list_dfleagues):
    
    l=[list_dfleagues[i].drop(["season","team_api_id"],1).mean().to_frame().transpose() for i in range(len(list_dfleagues))]
    df = pd.concat(l,ignore_index=True)
    
    return df
```

```python
df_leagues = leagues_to_df(list_dfleagues)
df_leagues[0:5]
```

```python
df = cl.normalisation(df_leagues)
M_fusion = cl.clustering_hierarchique_max2(df.as_matrix(),"manhattan",names_league)
```

```python
plt.figure(figsize = (50,15))
plt.title("Similarités entre les ligues", fontsize = 80)

scipy.cluster.hierarchy.dendrogram(M_fusion, labels=names_league, leaf_font_size=38, leaf_rotation=30)
plt.show()
```
